```python
import pandas as pd
import numpy as np
import re
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('raw_data.csv')
```

```python
df.head()
```

```python
df.shape
```

```python
# Won primetime emmys
df['emmy_won'] = df['Oscar'].apply(lambda x: int(''.join(filter(str.isdigit, str(x)))) if 'Won' in str(x) and ('Primetime Emmy' in str(x) or 'Primetime Emmys' in str(x)) else 0)

# Nominated for primetime emmys
df['emmy_nomination'] = df['Oscar'].apply(lambda x: int(''.join(filter(str.isdigit, str(x)))) if 'Nominated' in str(x) and ('Primetime Emmy' in str(x) or 'Primetime Emmys' in str(x)) else 0)

# nominated for BAFTA awards
df['bafta_nominated'] = df['Oscar'].apply(lambda x: int(''.join(filter(str.isdigit, str(x)))) if 'Nominated' in str(x) and ('BAFTA Award' in str(x) or 'BAFTA Awards' in str(x)) else 0)

# won BAFTA awards
df['bafta_won'] = df['Oscar'].apply(lambda x: int(''.join(filter(str.isdigit, str(x)))) if 'Won' in str(x) and ('BAFTA Award' in str(x) or 'BAFTA Awards' in str(x)) else 0)

# oscar won
df['oscar_won'] = df['Oscar'].apply(lambda x: int(''.join(filter(str.isdigit, str(x)))) if 'Won' in str(x) and ('Oscar' in str(x) or 'Oscars' in str(x)) else 0)

# oscar nomination
df['oscar_nomination'] = df['Oscar'].apply(lambda x: int(''.join(filter(str.isdigit, str(x)))) if 'Nominated for' in str(x) and ('Oscar' in str(x) or 'Oscars' in str(x)) else 0)
```

```python
df.head()
```

```python
# release date
df['Release Date'] = df['Release Date'].apply(lambda x: re.sub(r"[\[\]']","",x).split(',')[0].strip() if isinstance(x, str) else '')
```

```python
# extract wins  from award column
def extract_wins(text):
    wins_match = re.search(r'(\d+) win(s?)', text)
    if wins_match:
        return int(wins_match.group(1))
    else:
        return 0

# extract nomination  from award column
def extract_nominations(text):
    nominations_match = re.search(r'(\d+) nomination(s?)', text)
    if nominations_match:
        return int(nominations_match.group(1))
    else:
        return 0
        
# awards won
df['other_awards_won'] = df['Awards'].apply(extract_nominations)

# awards nomination 
df['other_award_nomiantion'] = df['Awards'].apply(extract_wins)
```

```python
df = df.drop(['Oscar', 'Awards'], axis=1)
```
